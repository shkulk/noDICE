```python
import numpy as np
import seaborn as sns
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-deep')
```

```python
exp = []
alg = []
tol2009 = []
tol2014 = []
nordhaus = []
weitzman = []
newboldmartin2014 = []

for i in range(0,21):
    exp.append(1-(np.exp(-0.0025*i**2.45)))
    alg.append(1-1/(1 + 0.001296610777435434*i**2.8871903728601973))
    nordhaus.append(0.00267*(i**2))
    if i<=3:
        newboldmartin2014.append(1.1*i/100)
    else:
        newboldmartin2014.append((1.1*i+2.5*(i-3))/100)
    #tol2009.append((-2.46*i+1.1*i**2)/100)
    tol2014.append((0.28*i + 0.16*i**2)/100)
    weitzman.append(1-1/(1+0.0028388**2+0.0000050703*(i**6.754)))
```

```python
sns.set("paper", style = "whitegrid", font_scale=1.75)
fig = plt.figure(figsize=(14,7))
plt.plot(np.linspace(0,20,21), alg, linewidth=4.0, color = sns.color_palette("viridis", 10)[3], label='Newbold & Daigeneault (2009) - Algebraic')
plt.plot(np.linspace(0,20,21), exp, linewidth=4.0, color = sns.color_palette("viridis", 10)[-4], label='Newbold & Daigeneault (2009) - Exponential')
#plt.plot(np.linspace(0,20,21), tol2009, linewidth=4.0, color = "forestgreen", label='Tol (2009)')
plt.plot(np.linspace(0,20,21), tol2014, color = "red", linewidth=4.0, label='Tol (2014)')
plt.plot(np.linspace(0,20,21), newboldmartin2014, color = "black", linewidth=4.0, label='Newbold & Martin (2014)')
plt.plot(np.linspace(0,20,21), nordhaus, linewidth=4.0, color = sns.color_palette("inferno", 15)[-3], label='Nordhaus (2013)')
plt.plot(np.linspace(0,20,21), weitzman, color = sns.cubehelix_palette(10)[5], linewidth=4.0, label='Weitzman (2010)')
plt.legend(fontsize = 15)
plt.xlabel("Temperature Change Relative to Pre-Industrial Level (°C)")
plt.ylabel("Damage as Fraction of Gross Output")
plt.show()
fig.savefig('Damage_Function.png', dpi=fig.dpi,  bbox_inches = "tight")
```
